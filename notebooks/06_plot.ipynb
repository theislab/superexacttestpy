## Functions

### color_map_color

```python
stest.pl.color_map_color(10,cmap_name='YlOrRd',vmin=0,vmax=100)
```

### Plotting function

#### Default parameters

```python
stest.pl.plot(data,total,data_name,size=(10,6))
```

You can get the plot and the df of results 

#### Custom parameters

##### Degree 

e.g with degree = 2, we choose only the intersection between 2 sets 

```python
tmp = stest.pl.plot(data,total,data_name,degree=2,size=(10,6))
```

##### Sort_by

```python
tmp = stest.pl.plot(data,total,data_name,sort_by="cardinality",size=(10,6))
```

##### Show count 

```python
tmp = stest.pl.plot(data,total,data_name,show_count=False,size=(10,6))
```

##### Orientation 

```python
tmp = stest.pl.plot(data,total,data_name,orientation="vertical",size=(10,6))
```

#### Background color 

```python
tmp = stest.pl.plot(data,total,data_name,degree=-1,background_color="dark_background",size=(10,6))
```
